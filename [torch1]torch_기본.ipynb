```python
import torch
x = torch.ones(2,2, requires_grad=True) # x 2,2 1로 채워진 텐서인데 이 옵션을 주면 해당옵션이라는 x라는텐서에서 연산된 모든 것들을 추적하기 시작, x가 변환이 되었을 때 어떤 연산을 하는지 기억

y =  x + 1
z = 2 * y **2
r = z.mean()
print('Result :', r)
```

```python
print(f'x : {x}')
print(f'y : {y}')
print(f'z : {z}')
print(f'r : {r}')
# r은 x에 관한 식이다. r z y x
```

```python
# 역으로 식을 써내려가보자
# res = (z_1 + z_2 + z_3+ z_4)/4
#z_i - 2 * y_i **2
#z_i = 2(x_i+1)**2
# d(res)/dx_i = x_i + 1

r.backward() # r을 기준으로 역전파를 진행한다
print(x.grad) # 모든 x의 값이 1임으로 x.grad를 통해 나온 미분값은 모두 2

# x.grad 는 backward()가 선언 된 변수를 기준으로 미분을 한다
# 즉 d(res)/dx를 계산한다
# d(res)dx_i = x_i + 1
```

```python
x = torch.FloatTensor(range(5)).unsqueeze(1)
y = 2 * x + torch.rand(5,1)
# torch.rand(5,1) 정규분포 난수값으로 구성된 5x1 텐서 생성
num_features = x.shape[1] # 특성의 수 저장
```

```python
print (x.shape, y.shape)
```

```python
# dtype = torch,FloatTensor
w = torch.randn(num_features, 1, requires_grad=True)
# 선형식은 y = xw + b로 표현, 따라서 w는 5x1 데이터와 곱할 수 있어야 함
# 예측값이 1개로 나와야 함으로 ㅡ기가 1(피처수)x1(출력값 크기)인 배열로 정의
b = torch.randn(1, requires_grad=True)
# b는 모든 인스턴스에 동일한 숫자를 더해 주어야함으로 1인 텐서로 정의
```

```python
b
```

```python
learning_rate = 1e-3
optimizer = torch.optim.SGD([w,b], lr=learning_rate)
# torch으 ㅣ옵티마이저는 torch,optim로 정의 할 수 있다,
```

```python
print(w.shape, b.shape)
```

```python
loss_stack = []
for epoch in range(1001):
    optimizer.zero_grad() # 최적화는 계싼을 누적시키기 때문에 매 에폭마다 값을 초기화 해야한다.
    y_hat = torch.matmul(x ,w) + b # 회귀식 모델을 이용하여 예측값 산출, 랜덤으로 해서 엉망일거임
    loss = torch.mean((y_hat - y) **2) # 손실함수 MSE를 이용해 오차 계산, mse 사용

    loss.backward() # 역전파를 기준으로 손실함수로 정의
    optimizer.step() # 미리 정의한 optimizer를 이용해 최적화 시행
    loss_stack.append(loss.item()) # 손실함수 값을 저장(item() 을 사용하지 않으면 loss 텐서 전체를 저장)

    if epoch % 100 == 0:
        print(f'Epoch : {epoch}, Loss : {loss.item():.3f}')
```

```python
with torch.no_grad(): # torch.no_grad()는 graient 연산을 옵션을 끔
    y_hat = torch.matmul(x, w) + b # w는 2에 근사, b는 0에 근사
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(loss_stack)
plt.title('Loss')
plt.subplot(1,2,2)
plt.plot(x, y,'.b')
plt.plot(x, y_hat,'r-')
plt.legend(['ground truth', 'prediction'])
plt.title('Prediction')
plt.show()
```

```python
train_images = np.random.randint(256, size=(100, 32, 32, 3))/255 # 이미지 수 x 너비 x 높이 x 채널 수
train_labels = np.random.randint(2, size=(100, 1))
```

```python
# 전처리 기술을 직접 만들어보자
# 기본 양식과 같이 ㅅ하용하기 위해 call 함수르 랏용한다
# def_call 내의 원하는 전처리 작업을 프로그래밍 할 수 있다
```

```python
class ToTensor:
  def __call__(self, sample):
    inputs, labels = sample # 입력된 sample 튜플에서 inputs와 labels를 각각 분리
    inputs = torch.FloatTensor(inputs) # 입력 데이터가 float형 텐서로 변환
    inputs = inputs.permute(2,0,1) # 채널 높이 너비
    return inputs, torch.LongTensor(labels)

# CutOut :
class CutOut:
    def __init__(self, ratio = 0.5): # CutOut이 적용될 확률을 조정
        self.ratio = int(1/ratio) # ratio 가 0.5이면 50% 확률로 CutOut 발현
    def __call__(self, sample):
        inputs, labels = sample
        active = int(np.random.randint(0, self.ratio,1 )) # 0 또는 1 반환 받음
        if active == 0:
            _, w, h = inputs.size() # .size()이미지의 너비와 높이를 받음 input 크기는 (채널, 너비 높이)
            min_len = min(w,h) # 최소값구하기
            box_size = int(min_len//4) # CutOut의 길이를 최소값의 1/4로 결정
            idx = int(np.random.randint(0, min_len-box_size, 1)) # idx를 통해 CutOut 박스좌측 상단 꼭지점 위치를
            inputs[:, idx:idx+box_size, idx:idx+box_size] = 0 # 해당 영역의 값 0으로 대체
            #  해당 영역의 값을 0으로 설정하여 잘라낸 효과를 만듭니다. 이미지 데이터에서 해당 영역이 검은색(0)으로 변하게 됩니다.
        return inputs, labels
```

```python
class MyDataset(Dataset):
  def __init__(self, x_data, y_data, transform=None):
    self.x_data = x_data # 입력 데이터(예: 이미지).
    self.y_data = y_data # 라벨 데이터(예: 이미지의 클래스).
    self.transform = transform
    self.len = len(y_data)
    self.tensor = ToTensor()

  def __getitem__(self, index):
    sample = self.x_data[index], self.y_data[index]
    if self.transform:
      sample = self.transform(sample)
    else:
      sample = self.tensor(sample)
    return sample

  def __len__(self):
    return self.len
```

```python
# CutOut :
class CutOut:
    def __init__(self, ratio = .5): # CutOut이 적용될 확률을 조정
        self.ratio = int(1/ratio) # ratio 가 0.5이면 50% 확률로 CutOut 발현
    def __call__(self, inputs):

        active = int(np.random.randint(0, self.ratio,1 )) # 0 또는 1 반환 받음
        if active == 0:
            _, w, h = inputs.size() # .size()이미지의 너비와 높이를 받음 input 크기는 (채널, 너비 높이)
            min_len = min(w,h) # 최소값구하기
            box_size = int(min_len//4) # CutOut의 길이를 최소값의 1/4로 결정
            idx = int(np.random.randint(0, min_len-box_size, 1)) # idx를 통해 CutOut 박스좌측 상단 꼭지점 위치를
            inputs[:, idx:idx+box_size, idx:idx+box_size] = 0 # 해당 영역의 값 0으로 대체
            #  해당 영역의 값을 0으로 설정하여 잘라낸 효과를 만듭니다. 이미지 데이터에서 해당 영역이 검은색(0)으로 변하게 됩니다.
        return inputs
```

```python
class MyDataset(Dataset):
  def __init__(self, x_data, y_data, transform=None):
    self.x_data = x_data # 입력 데이터(예: 이미지).
    self.y_data = y_data # 라벨 데이터(예: 이미지의 클래스).
    self.transform = transform # mytransform 규칙이 들어가게 됨
    self.len = len(y_data)
    self.tensor = ToTensor()

  def __getitem__(self, index):
    sample = self.x_data[index], self.y_data[index]
    if self.transform:
      sample = self.transform(sample)
    else:
      sample = self.tensor(sample)
    return sample

  def __len__(self):
    return self.len
```

```python
class MyTransform:
  def __call__(self, sample):
    inputs, labels = sample
    inputs = torch.FloatTensor(inputs)
    inputs = inputs.permute(2,0,1)
    labels = torch.FloatTensor(labels)

    transf = tr.Compose([tr.ToPILImage(), tr.Resize(128), tr.ToTensor(), CutOut()]) # 4가지를 모두 수행 후 final_output에 들어감
    final_output = transf(inputs)
    return final_output, labels
```
